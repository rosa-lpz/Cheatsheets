````python
import pandas as pd
````

# Data

````python
# initialize list of lists
data = [['1949', 'Jan', 112, 48], ['1949', 'Feb', 118, 50], ['1949', 'Mar', 132, 54], ['1949', 'Apr', 129, 51], ['1949', 'May', 121, 55]]

# Create the pandas DataFrame
flights = pd.DataFrame(data, columns=['year', 'month','passengers', 'ticket_cost (USD)'])

# print dataframe.
flights.head()
````

# Method 1: Create a new column inside a DataFrame

````python
flights['total_cost'] = flights['passengers'] * flights['ticket_cost (USD)']
flights
````

# Method 2: Add a column in a DF using assing function



```
# Syntax of DataFrame.assign()
DataFrame.assign(**kwargs)
```



````python
totalCost=[5376,5900,7128,6579,6655]
flights2 = flights.assign(total_cost=totalCost)
flights2
````

# Add multiple columns in a DF using assign function

````python
#totalCost=[5376,5900,7128,6579,6655]
flightNumber=['A34','B12','C43','D14','E29']
flights3 = flights.assign(total_cost=totalCost, flight_number=flightNumber)
flights3
````

# Add a Constant or Empty columns


````python
# Add a constant or empty value to the DataFrame.
df_constant=flights.assign(A=None,B=0,C="")
df_constant
````
