```python
# Option 1 - Get the list of all column names from headers
column_names = list(flights.columns.values)
column_names
```

```python
# Option 2 - Get the list of all column names from headers
column_names2 = flights.columns.values.tolist()
column_names2
```

```python
# Option 3 - Using list(df) to get the column headers as a list
column_names3 = list(flights.columns)
column_names3
```

```python
# Option 4 - Using list(df) to get the list of all Column Names
column_names4 = list(flights)
column_names4
```

```python
# Option 5 - Dataframe show all columns sorted list
column_names5=sorted(flights)
column_names5
```

```python
# Option 6 - Get all Column Header Labels as List
for column_headers6 in flights.columns:
    print(column_headers6)

column_names6 = flights.keys().values.tolist()
```
