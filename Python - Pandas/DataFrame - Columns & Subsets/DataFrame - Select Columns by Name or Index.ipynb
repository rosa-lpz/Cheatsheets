# Select Columns by Name using df[] Notation

````python
# By using df[] Notation
df1 = flights[["year","passengers"]] # select multile columns
df1.head(10)
````

# Select Columns Using loc[] to take column slices

By using pandas.DataFrame.loc[] you can select columns by names or labels. To select the columns by names, the syntax is df.loc[:,start:stop:step]; where start is the name of the first column to take, stop is the name of the last column to take, and step as the number of indices to advance after each extraction; for example, you can select alternate columns. Or, use the syntax: [:,[labels]] with labels as a list of column names to take.



```
#loc[] syntax to slice columns
df.loc[:,start:stop:step]
```



````python
df2_1 = flights.loc[:, ["year","passengers"]] # Select multiple columns
df2_1.head(10)
````

````python
df2_2=flights.iloc[:2]
df2_2
````

## Select columns between two columns

````python
df3_1 = flights.loc[:,'month':'ticket_cost']
df3_1.head(10)
````

## Select columns by range

````python
df3_3 = flights.loc[:,:'passengers']  # Select columns by range
df3_3.head(10)
````

# Select Columns using iloc[] to select column by Index



```
df2 = df.iloc[:,[1,3,4]] # Select columns by Index
df2 = df.iloc[:,1:4] # Select between indexes 1 and 4 (2,3,4)
df2 = df.iloc[:,2:] # Select From 3rd to end
df2 = df.iloc[:,:2] # Select First Two Columns
```



## Select columns by index

````python
df4_1 = flights.iloc[:,[1,3]] # Select columns by Index
df4_1.head(10)
````

## Select columns by index range

````python
df4_2 = flights.iloc[:,1:4] # Select between indexes 1 and 4 (2,3,4)
df4_2
````
