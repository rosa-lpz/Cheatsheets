The np-dot-sort function sorts an array along a given axis. Here, NumPy has sorted the array along the columns, with the highest number from each row in the far right.

# Slicing 2D Arrays

## Sudoku game

For example, to get the value in the third row of the fifth column, we provide both the row index, two, and the column index, four. Here, like everywhere else in NumPy, row information comes before column information.

```python
# Example 2
sudoku_array = np.array([
 [0, 0, 4, 3, 0, 0, 2, 0, 9],
 [0, 0, 5, 0, 0, 9, 0, 0, 1],
 [0, 7, 0, 0, 6, 0, 0, 4, 3],
 [0, 0, 6, 0, 0, 2, 0, 8, 7],
 [1, 9, 0, 0, 0, 7, 4, 0, 0],
 [0, 5, 0, 0, 8, 3, 0, 0, 0],
 [6, 0, 0, 0, 0, 0, 1, 0, 5],
 [0, 0, 3, 5, 0, 8, 6, 9, 0],
 [0, 4, 2, 9, 1, 0, 3, 0, 0]])

sudoku_array
```

### Indexing

```python
sudoku_array[2,4]
```

```python
# Index the first row
sudoku_array[0]
```

```python
# Index the fourth colum
sudoku_array[:,3]
```

### Slicing

```python
sudoku_array[3:6, 3:6]
```

```python
# Slicing with steps
sudoku_array[3:6:2, 3:6:2]
```

### Sorting

The np-dot-sort function sorts an array along a given axis. Here, NumPy has sorted the array along the columns (axis=1), with the highest number from each row in the far right.

The default axis in np-dot-sort is the last axis of the array passed to it. If a 2D array is being sorted, NumPy sorts by columns, since columns are axis 1 and rows are axis 0.


```python
np.sort(sudoku_array)
```

### Sorting along the rows

```python
np.sort(sudoku_array, axis=0)
```
