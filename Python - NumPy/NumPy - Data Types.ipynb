```python
import numpy as np
```

```python
arr1 = np.array([1,2,3,4])
arr2 = np.array(['apple','banana','cherry'])

print(arr1.dtype)
print(arr2.dtype)
```

```python
# Create an array with data type string:
arr = np.array([1, 2, 3, 4], dtype='S')

print(arr)
print(arr.dtype)
```

```python
# Create an array with data type 4 bytes integer:
arr = np.array([1, 2, 3, 4], dtype='i4')

print(arr)
print(arr.dtype)
```
